# **Jogando 21 com agente de políticas fixas**

```python
import gymnasium as gym
import time
```

```python
env = gym.make("Blackjack-v1", render_mode="human")
```

## **Definição da política (regra) a ser usada**

```python
def fixed_policy(obs):
    player_sum, dealer_card, usable_ace = obs

    if player_sum >= 18:
        return 0 # não pede mais uma carta (stick)
    else:
        return 1 # pede mais uma carta (hit)
```

## **Aplicando a política em um episódio apenas**

```python
def play_episode(env, policy):
    obs, info = env.reset() # resetando o ambiente e obtendo a primeira observação de seu estado

    total_reward = 0.0
    steps = 0

    while True:
        time.sleep(5)
        action = policy(obs) # escolhendo uma ação de acordo com a política e a observação do ambiente
        obs, reward, terminated, truncated, info = env.step(action) # performando a ação escolhida
        time.sleep(5)
        total_reward += reward
        steps += 1

        if terminated or truncated:
            print(f"Fim do episódio. Recompensa total: {total_reward}")
            break

    return total_reward
```

```python
episode = play_episode(env, fixed_policy)
time.sleep(5)

env.close() # precisamos fechar o ambiente depois de rodar todos os episódios (ciclos de ação)
```

```python
def play_multi_episodes(env, policy, n_episodes):
    wins = draws = losses = 0
    total_reward = 0.0

    for i in range(n_episodes):
        print(i)
        time.sleep(3)
        obs, info = env.reset()

        while True:
            time.sleep(3)
            action = policy(obs)
            obs, reward, terminated, truncated, info = env.step(action)
            time.sleep(3)

            total_reward += reward

            if terminated or truncated:
                if reward > 0:
                    wins += 1
                elif reward < 0:
                    losses += 0
                else:
                    draws += 1
                break
    return n_episodes, wins, draws, losses, (total_reward/n_episodes)
```

```python
episodes, wins, draws, losses, avg_reward = play_multi_episodes(env, fixed_policy, 3)
# print(f"Quantidade de episódios rodados: {episodes}\nVitórias: {wins}\nEmpates: {draws}\nDerrotas: {losses}\nMédia de recompensa: {avg_reward:.2f}")

env.close()
```
