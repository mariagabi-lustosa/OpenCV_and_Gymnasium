# **Frozen Lake com agente de políticas fixas**

````python
import gymnasium as gym
#import time


# definindo um dicionário para as ações, usando números como chaves
ACTIONS = {0: "Left",
           1: "Down",
           2: "Right",
           3: "Up"}

env = gym.make("FrozenLake-v1", desc=None, render_mode="human", is_slippery=False, map_name="4x4")
# desc=["SFFF", "FHFH", "FFFH", "HFFG"]
n_cols = 4

#print(env.render())
````

### **Observation Space**
A observação no caso desse ambiente é um valor que representa a posição atual do jogador como current_row * n_cols + current_col (onde tanto a linha quanto a coluna começam em 0).

Por exemplo, a posição do prêmio no mapa 4x4 pode ser calculada da seguinte forma: 3*4+3=15. Assim, o número de observações possíveis depende do tamanho do mapa.

````python
print(env.observation_space)
````

### **Definição da política**
Vamos usar um conjunto de regras muito simples: se o índice da coluna atual em que o agente está for menor que a quantidade de colunas do tabuleiro menos 1, o agente deve ir para a direita. Caso contrário, deve andar para baixo.

```obs``` é a posição atual do agente no tabuleiro. O operador % (módulo, resto da divisão) está sendo usado para descobrir a coluna onde o agente está. Ex.: ```obs``` = 6 -> 6 % 4 = 2, ou seja, está na coluna 2 da segunda linha.

````python
def policy(obs, n_cols):
    col = obs % n_cols
    if col < n_cols - 1:
        return 2 # right
    return 1 # down
````

### **Executando os episódios**

````python
def run_episode(env, policy, total_reward):
    obs, info = env.reset()
    total_reward = 0.0

    for i in range(50):  # limite de passos arbitrário
        action = policy(obs, n_cols)  # usa a política fixa
        obs, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

        print(f"Passo {i+1} | Ação: {ACTIONS[action]}")
        #print(env.render())
        #time.sleep(0.2)

        if terminated or truncated:
            if reward > 0:
                print("Sucesso! Chegou ao objetivo.")
            else:
                print("Fim do episódio (buraco ou truncado).")
            break
    return total_reward
````

````python
episode = 0
for i in range(5):
    episode += run_episode(env, policy, episode)
    print(f"Recompensa total: {episode}")

env.close()
````
