# **Introdução ao Gymnasium**
Vamos começar fazendo um tour pelas funções básicas da biblioteca e que permitem implementar as interações agente-ambiente.

### **gym.make(env_name, render_mode, ...)**
Essa função cria o ambiente e a sua interface.

````python
import gymnasium as gym
env = gym.make('Blackjack-v1', natural=False, sab=False, render_mode="human")
````

### **env.reset()**
Reinicia o ambiente para um estado inicial. Ela marca o início de cada episódio.
- Retorna: ```(obs, info)```

```obs``` é a observação (estado inicial) e info são metadados extras (dependem do ambiente).

````python
observation, info = env.reset()
print(observation)
````

### **env.action_space() e env.observation_space()**
Descrevem o espaço de ações e observações possíveis.

Os ambientes sempre tem um agente. É ele que irá observar o estado atual do ambiente e escolher uma ação (que pode ou não levar em conta essas informações disponíveis). O agente pode ir aprendendo as regras (políticas) do ambiente ao longo do tempo de acordo com as recompensas recebidas, ou apenas executar regras fixas.

````python
print(env.action_space)       # Discrete(2) -> 0: stick, 1: hit
print(env.observation_space)  # Tuple(Discrete(32) -> soma atual do player, Discrete(11) -> valor da carta mostrada pelo dealer, Discrete(2) -> se player seguraum ás usável)
````

### **env.step(action)**
Aplica uma ação no ambiente e avança um passo no tempo.
- Retorna: ```obs```, ```reward```, ```terminated```, ```truncated```, ```info```.
```obs``` é a próxima observação, ```reward``` é a recompensa imediata, ```terminated``` é um booleano que é True se episódio terminou de forma natural (chegou ao fim do jogo, o agente caiu, etc.), ```truncated``` é outro booleano que é True se episódio acabou por limite externo (ex: max steps). Por fim, ```info``` são extras (depuração, métricas, etc.).

````python
episode_over = False
total_reward = 0

while not episode_over:
    action = env.action_space.sample()

    observation, reward, terminated, truncated, info = env.step(action)

    total_reward += reward
    episode_over = terminated or truncated
````

### **env.close()**
Libera os recursos como janelas gráficas, processos de simulação, dentre outros.

````python
print (total_reward)
env.close()
````

## **Exemplo de ambiente implementado**
Abaixo vamos fazer a implementação mais simples possível do Blackjack, que é equivalente ao nosso jogo "21" no Brasil.

````python
import gymnasium as gym

env = gym.make('Blackjack-v1', render_mode="human")
obs, info = env.reset()

done = False
while not done:
    action = env.action_space.sample()  # agente aleatório
    obs, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated

env.close()
````
